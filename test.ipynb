```python
import pandas as pd
from scia import plot
import matplotlib.pyplot as plt

# Your data
df = pd.DataFrame({
    'mt': range(1, 13),
    'values': [50, 58, 75, 63, 70, 59, 64, 69, 72, 77, 76, 73],
    'phase': ['A1']*3 + ['B1']*3 + ['A2']*3 + ['B2']*3
})

# Basic plot
plot(df)
plt.show()

# Custom colors
colors = {
    'A1': 'lightblue',
    'B1': 'lightgreen',
    'A2': 'skyblue',
    'B2': 'lightseagreen'
}

# Plot with custom colors
plot(df, colors=colors, title='My Single-Case Data')
plt.show()
```

```python
import pandas as pd
from scia import plot
import matplotlib.pyplot as plt

# Create multiple case data
# Case 1: Tick
tick_data = pd.DataFrame({
    'mt': range(1, 13),
    'values': [50, 58, 75, 63, 70, 59, 64, 69, 72, 77, 76, 73],
    'phase': ['A1']*3 + ['B1']*3 + ['A2']*3 + ['B2']*3
})

# Case 2: Trick
trick_data = pd.DataFrame({
    'mt': range(1, 13),
    'values': [51, 59, 70, 65, 68, 64, 65, 70, 70, 68, 69, 70],
    'phase': ['A1']*3 + ['B1']*3 + ['A2']*3 + ['B2']*3
})

# Case 3: Track
track_data = pd.DataFrame({
    'mt': range(1, 13),
    'values': [54, 57, 51, 61, 57, 53, 58, 57, 60, 55, 58, 58],
    'phase': ['A1']*3 + ['B1']*3 + ['A2']*3 + ['B2']*3
})

# Combine all cases into a list
cases = [tick_data, trick_data, track_data]

# Custom colors for phases
colors = {
    'A1': 'lightblue',
    'B1': 'lightgreen',
    'A2': 'skyblue',
    'B2': 'lightseagreen'
}

# Plot all cases with custom colors
plot(cases, colors=colors, title='Multiple Cases Example', 
     ylabel='Score', figsize=(12, 10))
plt.show()
```
